```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv('Asthma_backup.csv')
print(df.columns)
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
df.head(10)
```

```python
bins = [0,99,125]
group_names =['0','1']
df['output1'] = pd.cut(df['heart_pulse'],bins,labels = group_names)
df = df.dropna(subset=['output1'])
df['output1'] = df['output1'].astype(int)
```

```python
df.head(5)
```

```python
bins = [0,92,95]
group_names =['0','1']
df['output2'] = pd.cut(df['oxygen_saturation'],bins,labels = group_names)
df = df.dropna(subset=['output2'])
df['output2'] = df['output2'].astype(int)
```

```python
df.head(3)
```

```python
bins = [0,20,30]
group_names =['0','1']
df['output3'] = pd.cut(df['breaths_per_minute'],bins,labels = group_names)
df = df.dropna(subset=['output3'])
df['output3'] = df['output3'].astype(int)
```

```python
df.head(3)
```

```python
df.head(5)
```

```python
df['class'] = np.where( ( (df['output1'] ) & (df['output2']) & (  df['output3'])  ) , 1, 0)
```

```python
df
```

```python
x=df[["heart_pulse","oxygen_saturation","breaths_per_minute"]]
y=df[["class"]]


```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size =0.25,random_state=66)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
import sklearn
from sklearn.neural_network import MLPClassifier
```

```python
clf = MLPClassifier(hidden_layer_sizes=(51),solver="lbfgs",alpha=1e-5)
```

```python
clf.fit(x,y)
```

```python
y_pred = clf.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_train,y_pred))
```
